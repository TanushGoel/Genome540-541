```python
import pandas as pd

metadata = pd.read_csv("sars2-metadata.csv")
metadata.head()
```

```python
with open("sars2.fasta.contree", 'rt') as f:
    data = f.read()
```

```python
for strain in metadata["strain"]:
    data = data.replace(strain, metadata[metadata["strain"] == strain]["annotation"].iloc[0].upper() + "_" + strain)
```

```python
with open("updated_sars2.fasta.contree", "w") as f:
    f.write(data)
f.close()
```
